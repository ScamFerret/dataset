```python
# Loading python modules
import glob
import json
import collections
```

```python
# Table 5: Results of Groundedness Evaluation per LLM
path_list = glob.glob('./gpt-4_results/explainability/*/*/*.json')
result_list = list()
for path in path_list:
    with open(path, 'r') as file:
        data = json.load(file)
    result_list.append(data['result'])
c = collections.Counter(result_list)
custom_order = ['Very good', 'Good', 'Okay', 'Somewhat bad', 'Very bad']
sorted_data = sorted(c.most_common(), key=lambda x: custom_order.index(x[0]))
print ('----------GPT-4----------')
print (sorted_data)

path_list = glob.glob('./gpt-3.5_results/explainability/*/*/*.json')
result_list = list()
for path in path_list:
    with open(path, 'r') as file:
        data = json.load(file)
    result_list.append(data['result'])
c = collections.Counter(result_list)
custom_order = ['Very good', 'Good', 'Okay', 'Somewhat bad', 'Very bad']
sorted_data = sorted(c.most_common(), key=lambda x: custom_order.index(x[0]))
print ('----------GPT-3.5----------')
print (sorted_data)

path_list = glob.glob('./geminipro_results/explainability/*/*/*.json')
result_list = list()
for path in path_list:
    with open(path, 'r') as file:
        data = json.load(file)
    result_list.append(data['result'])
c = collections.Counter(result_list)
custom_order = ['Very good', 'Good', 'Okay', 'Somewhat bad', 'Very bad']
sorted_data = sorted(c.most_common(), key=lambda x: custom_order.index(x[0]))
print ('----------Gemini Pro----------')
print (sorted_data)
```
